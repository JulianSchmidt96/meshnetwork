```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
floor_small_distance = pd.read_csv('iperf_log_router-3_floor_meas-01.csv')
floor_large_distance = pd.read_csv('iperf_log_router-3_floor_meas-02.csv')
```

```python
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(floor_small_distance['datarate'], label = 'small distance')
plt.plot(floor_large_distance['datarate'], label = 'large distance')
plt.legend()
plt.xlabel('time in seconds')
plt.ylabel('send data in bits')
plt.title('send data in overtime for different  distances')
plt.savefig('plots/send_data_in_overtime_for_different_distances.png')
```

```python
#plt.style.use('seaborn-deep')
plt.figure(figsize=(12,6))
plt.grid()
data = [floor_small_distance['datarate'], floor_large_distance['datarate']]
labels = ['small distance', 'large distance']
plt.hist(data, bins = 50, stacked=True)
plt.legend(labels)
plt.xlabel('time in seconds')
plt.ylabel('send data in bits')
plt.title('send data  for different  distances')
plt.tight_layout()
plt.savefig('plots/send_data_hist_for_different_distances.png')
```
