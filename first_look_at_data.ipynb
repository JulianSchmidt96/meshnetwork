```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
floor_small_distance = pd.read_csv('iperf_log_router-3_floor_meas-01.csv')
floor_large_distance = pd.read_csv('iperf_log_router-3_floor_meas-02.csv')
```

```python
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(floor_small_distance['data'], label = 'small distance')
plt.plot(floor_large_distance['data'], label = 'large distance')
plt.legend()
plt.xlabel('time in seconds')
plt.ylabel('send data in bits')
plt.title('send data in overtime for different  distances')
plt.savefig('plots/send_data_in_overtime_for_different_distances.png')
```

```python
floor_measuremt_router_3 = pd.read_csv( 'iperf_log_router-3_floor_2clients_meas_01.csv')
floor_measuremt_router_4 = pd.read_csv( 'iperf_log_router-4_floor_2clients_meas_01.csv')
```

```python
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(floor_measuremt_router_3['data'], label = 'router-3')
plt.plot(floor_measuremt_router_4['data'], label = 'router-4')
plt.legend()
plt.xlabel('time in seconds')
plt.ylabel('send data in bits')
plt.title('send data in overtime for measurement with two clients')
plt.savefig('plots/send_data_in_overtime_for_measurement_with_two_clients.png')
```
